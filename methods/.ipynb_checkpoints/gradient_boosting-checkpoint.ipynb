```python
import numpy as np
import os
os.chdir("..")
data = np.loadtxt(('data/preprocessed-dataset.csv'), delimiter=',', skiprows=1)
os.chdir("methods")
x = data[:,1:35]
y = data[:,35]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1, test_size=0.2)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc_model=GradientBoostingClassifier()
gbc_model.fit(x_train, y_train)

```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

y_predict=gbc_model.predict(x_test)

print('Accuracy: {}'.format(accuracy_score(y_test, y_predict)))
print('Precision: {}'.format(precision_score(y_test, y_predict)))
print('Recall: {}'.format(recall_score(y_test, y_predict)))
```

OPTIMIZATION

```python
import numpy as np
n_estimators=[int(x) for x in np.linspace(start=10, stop=100, num=10)]
learning_rate = [0.01, 0.1, 0.5, 1]
min_samples_split = [2,5]
min_samples_leaf = [1,2]
max_depth = [2,4]
max_features = ['sqrt', None]

param_grid={'n_estimators': n_estimators, 'learning_rate' : learning_rate, 'min_samples_split' : min_samples_split, 'min_samples_leaf' : min_samples_leaf, 'max_depth' : max_depth, 'max_features' : max_features}
```

```python
gb_model=GradientBoostingClassifier()
GradientBoostingClassifier().get_params().keys()
```

```python
from sklearn.model_selection import GridSearchCV
gb_grid=GridSearchCV(gb_model, param_grid=param_grid, cv=10, verbose=2, n_jobs=4)
gb_grid.fit(x_train, y_train)
```

```python
gb_grid.best_params_
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, roc_auc_score
gb_model=GradientBoostingClassifier(learning_rate=0.1, max_depth=4, max_features='sqrt', min_samples_leaf=2, min_samples_split=2, n_estimators=70)
gb_model.fit(x_train, y_train)
y_predict=gb_model.predict(x_test)

print('Accuracy: {}'.format(accuracy_score(y_test, y_predict)))
print('Precision: {}'.format(precision_score(y_test, y_predict)))
print('Recall: {}'.format(recall_score(y_test, y_predict)))
print(confusion_matrix(y_test, y_predict))
print(roc_auc_score(y_test, y_predict))
```
